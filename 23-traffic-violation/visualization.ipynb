### 1. Which colors of the vehicles are more likely to get involved in a traffic violation?

```python
# Set size of plots
sns.set(rc={'figure.figsize':(15,10)})
```

```python
# Count values using pands
df["Color"].value_counts()[:5]
```

Black cars are the easiest to get involved in violation

```python
# Plot using seaborn and order using it's count
sns.countplot(y="Color", data=df, order=df["Color"].value_counts().index)
```

```python
sns.countplot(y="Color", hue="Gender", data=df, order=df["Color"].value_counts().index)
```

```python
sns.countplot(y="Color", hue="Race", data=df, order=df["Color"].value_counts().index)
```

#### Black Cars are the easiest one to get into traffic violation
